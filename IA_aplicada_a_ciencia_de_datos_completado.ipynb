> #### <strong><font color="7b71ff" size=6>Prompt:</font></strong>

Tenemos un DataFrame llamado "df" y queremos construir un gráfico de barras horizontales que muestre la facturación por categoría de Zoop.

En "df", tenemos las siguientes columnas:
* categoria: categorías de los productos
* precio_unitario: precio de una unidad del producto
* cantidad: cantidad vendida del producto
* envio: costo de envío total de la compra.

Deseamos crear una visualización que muestre la facturación de cada categoría, donde la facturación se calcula multiplicando el precio unitario por la cantidad y sumando el costo de envío.

Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.


````python
# Calcular la facturación para cada fila
df['facturacion_producto'] = df['precio_unitario'] * df['cantidad']
df['facturacion_total'] = df['facturacion_producto'] + df['envio']

# Agrupar por categoría y sumar la facturación total
facturacion_por_categoria = df.groupby('categoria')['facturacion_total'].sum().reset_index()

# Ordenar por facturación descendente para una mejor visualización
facturacion_por_categoria = facturacion_por_categoria.sort_values(by='facturacion_total', ascending=False)

facturacion_por_categoria
````

````python
# Crear la visualización de barras horizontales
plt.figure(figsize=(10, 8))
sns.barplot(x='facturacion_total', y='categoria', data=facturacion_por_categoria, palette='viridis')
plt.title('Facturación por Categoría de Producto en 2023')
plt.xlabel('Facturación Total (MXN)')
plt.ylabel('Categoría de Producto')
plt.tight_layout()
plt.show()
````

Tenemos un DataFrame llamado "df" y queremos construir un gráfico de líneas que muestre las ventas totales mensuales de Zoop.

En "df", tenemos las siguientes columnas:
* fecha: con las fechas de las compras en formato datetime (`aaaa-mm-dd`)
* facturación: calculada como el producto del precio unitario y la cantidad, sumado al costo de envío de cada venta.

Deseamos crear una visualización que muestre las ventas por mes. Primero, agrupa los datos por mes y luego crea una nueva columna llamada "mes" que contenga el nombre de cada mes, traduciendo los meses del índice mediante el siguiente diccionario. Utiliza la columna "fecha" para agrupar los datos y construir el gráfico. La columna "mes" solo debe usarse para cambiar las etiquetas del eje x.

```
meses = {
    'January': 'Ene',
    'February': 'Feb',
    'March': 'Mar',

    'May': 'May',
    'June': 'Jun',

```


Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.


````python
# Asegurarse de que la columna 'fecha' esté en formato datetime
df['fecha'] = pd.to_datetime(df['fecha'])

# Calcular la facturación si aún no existe la columna 'facturacion'
if 'facturacion' not in df.columns:
    df['facturacion'] = (df['precio_unitario'] * df['cantidad']) + df['envio']

# Extraer el mes de la fecha y agrupar por mes para sumar la facturación
ventas_mensuales = df.groupby(df['fecha'].dt.to_period('M'))['facturacion'].sum().reset_index()

# Convertir el período a string para mapear los nombres de los meses
ventas_mensuales['fecha'] = ventas_mensuales['fecha'].astype(str)

# Mapear los nombres de los meses usando el diccionario proporcionado
meses = {
    '2023-01': 'Ene',
    '2023-02': 'Feb',
    '2023-03': 'Mar',
    '2023-04': 'Abr',
    '2023-05': 'May',
    '2023-06': 'Jun',
    '2023-07': 'Jul',
    '2023-08': 'Ago',
    '2023-09': 'Sep',
    '2023-10': 'Oct',
    '2023-11': 'Nov',
    '2023-12': 'Dic'
}

ventas_mensuales['mes'] = ventas_mensuales['fecha'].map(meses)

ventas_mensuales
````

##**2.3 - Componiendo datos dinámicos**

###**4a Pregunta:** ¿Cómo está dividida la venta por trimestre por método de pago? ¿Existe algún período en el que algún tipo de pago se destaque?


> #### <strong><font color="7b71ff" size=6>Prompt:</font></strong>

Tenemos un DataFrame llamado "df" y queremos construir un gráfico de barras apiladas que muestre las ventas por trimestre en relación con el método de pago de Zoop.

En "df", tenemos las siguientes columnas:
* fecha: con las fechas de las compras en formato datetime (`aaaa-mm-dd`)
* facturacion: con los valores de las ventas
* metodo_pago: con el método de pago elegido en la compra

Deseamos crear una visualización que muestre las ventas por trimestre, agrupando los métodos de pago de cada trimestre correspondientes a cada compra.

Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.


````python
# Asegurarse de que la columna 'fecha' esté en formato datetime
df['fecha'] = pd.to_datetime(df['fecha'])

# Calcular la facturación si aún no existe la columna 'facturacion'
if 'facturacion' not in df.columns:
    df['facturacion'] = (df['precio_unitario'] * df['cantidad']) + df['envio']

# Extraer el trimestre de la fecha
df['trimestre'] = df['fecha'].dt.to_period('Q')

# Agrupar por trimestre y método de pago, y sumar la facturación
ventas_trimestre_metodo = df.groupby(['trimestre', 'metodo_pago'])['facturacion'].sum().reset_index()

# Pivotear la tabla para que los métodos de pago sean columnas
ventas_trimestre_metodo_pivot = ventas_trimestre_metodo.pivot(index='trimestre', columns='metodo_pago', values='facturacion').fillna(0)

ventas_trimestre_metodo_pivot
````

````python
# Crear el gráfico de barras apiladas
ventas_trimestre_metodo_pivot.plot(kind='bar', stacked=True, figsize=(10, 7))
plt.title('Ventas por Trimestre y Método de Pago en 2023')
plt.xlabel('Trimestre')
plt.ylabel('Facturación Total (MXN)')
plt.xticks(rotation=0)
plt.legend(title='Método de Pago')
plt.tight_layout()
plt.show()
````

---

El gráfico de columnas apiladas añade a la comparación entre períodos o categorías, la comparación también entre los datos de cada período/categoría. Es como si pusiéramos una "lupa" en cada columna y visualizáramos la composición de cada una de acuerdo con la variable deseada (aquí, las formas de pago).

## **4.4 - Agregar elementos visuales**

---


¿Qué tal el uso de textos informativos, verdad? ¿Y si ahora añadimos otros elementos visuales a uno de nuestros gráficos?

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from matplotlib.ticker import FuncFormatter
from matplotlib.patches import Circle

# Asegúrate de que tus dataframes 'clientes' y 'ventas' están cargados
# y que el dataframe 'df' está unido y con las columnas requeridas.
# También asegúrate de que las variables de color (ROJO_1, AZUL_1, GRIS_1, etc.)
# y el diccionario 'meses' están definidos en tu entorno.

# Define tus variables de color y el diccionario 'meses' aquí si aún no lo has hecho
ROJO_1,	ROJO_2,	ROJO_3 = "#e23155",	"#cc2c4e", "#b32742"
AZUL_1,	AZUL_2,	AZUL_3 = "#203f75",	"#1c3867", "#19325b"
BLANCO,GRIS_1, GRIS_2, GRIS_3, GRIS_4,	GRIS_5 = "#ffffff", "#ebebeb",	"#d9d9d9", "#cccccc", "#555655", "#231f20"
AQUA_1,	AQUA_2,	AQUA_3 = "#addcd4",	"#9fccc5", "#96bfb9"

meses = {
    'January': 'Ene',
    'February': 'Feb',
    'March': 'Mar',
    'April': 'Abr',
    'May': 'May',
    'June': 'Jun',
    'July': 'Jul',
    'August': 'Ago',
    'September': 'Sep',
    'October': 'Oct',
    'November': 'Nov',
    'December': 'Dic'
}


# --- Funciones para generar cada gráfico ---

def metodo_pago():
    """Genera el gráfico de barras verticales de métodos de pago."""
    # Construir la tabla resumen de métodos de pago
    metodos_de_pago = df['metodo_pago'].value_counts().reset_index()
    metodos_de_pago.columns = ['Metodo_de_Pago', 'Cantidad']

    # Calcular el porcentaje de los dos primeros métodos de pago
    suma_dos_primeros = metodos_de_pago['Cantidad'].iloc[:2].sum()
    suma_total = metodos_de_pago['Cantidad'].sum()
    porcentaje = (suma_dos_primeros / suma_total) * 100

    # Crear la visualización
    fig, ax = plt.subplots(figsize=(10, 6), facecolor=GRIS_1)

    # Configurar el fondo del gráfico
    ax.set_facecolor(GRIS_1)

    # Crear el gráfico de barras verticales
    sns.barplot(x='Metodo_de_Pago', y='Cantidad', data=metodos_de_pago, color=ROJO_1, ax=ax)

    # Configurar el título
    ax.set_title('Métodos de pago más utilizados en 2023', fontsize=18, color=GRIS_5)

    # Configurar el eje x
    ax.set_xticks([])
    ax.set_xlabel('')

    # Configurar el eje y
    ax.tick_params(axis='y', labelsize=12, colors=AZUL_1)
    ax.set_ylabel('Cantidad', fontsize=12, color=AZUL_1)

    # Suprimir bordes del gráfico
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_color(GRIS_5)
    ax.spines['bottom'].set_visible(False)

    # Añadir etiquetas de texto con el total de cada método de pago encima de cada barra
    for p in ax.patches:
        height = p.get_height()
        plt.text(p.get_x() + p.get_width() / 2.,
                 height + 50,
                 '{:,.0f}'.format(height),
                 ha='center',
                 va='bottom',
                 color=AZUL_1,
                 fontsize=12)

    # Añadir el texto dentro del área del gráfico
    plt.text(0.5, 4000, # Ajusta estas coordenadas x e y
             f"{porcentaje:.2f}% de los clientes utilizan Tarjeta de Crédito o Transferencia \npara el pago. Sería interesante considerar asociaciones con \nbancos para la creación de un Zoop Pay con el fin de fidelizar \na nuestra clientela.",
             fontsize=10, color=AZUL_3,
             ha='center', va='top')

    plt.tight_layout()
    return fig # Devuelve el objeto figura


def ventas_mensuales():
    """Genera el gráfico de líneas de ventas mensuales."""
    # Asegurarse de que la columna 'fecha' esté en formato datetime
    df['fecha'] = pd.to_datetime(df['fecha'])

    # Calcular la facturación si aún no existe la columna 'facturacion'
    if 'facturacion' not in df.columns:
        df['facturacion'] = (df['precio_unitario'] * df['cantidad']) + df['envio']

    # Crear la columna 'mes' con el nombre completo del mes (probablemente en inglés)
    df['mes'] = df['fecha'].dt.strftime('%B')

    # Agrupar por mes y sumar la facturación, reindexando para asegurar el orden correcto de los meses
    ventas_mensuales_data = df.groupby('mes')['facturacion'].sum().reindex(meses.keys())

    # Crear la visualización
    fig, ax = plt.subplots(figsize=(12, 6), facecolor=GRIS_1)

    # Configurar el fondo del gráfico
    ax.set_facecolor(GRIS_1)

    # Crear el gráfico de líneas usando Matplotlib
    line, = ax.plot(range(len(ventas_mensuales_data.index)), ventas_mensuales_data.values, marker='o', color=AZUL_1, linewidth=2)

    # Obtener las coordenadas de los puntos de la línea (ahora x_data será numérico)
    x_data = line.get_xdata()
    y_data = line.get_ydata()

    # Definir los índices de los puntos donde dibujar los círculos
    indices_puntos = [4, 7, 10, 11] # Índices 0-based para 5º, 8º, 11º y 12º puntos

    # Dibujar los círculos y añadir la numeración
    for i, idx in enumerate(indices_puntos):
        centro_x = x_data[idx]
        centro_y = y_data[idx]

        circle = Circle((centro_x, centro_y), radius=0.3, # Radio ajustado para el eje x numérico (puedes necesitar probar valores)
                        edgecolor=ROJO_1, facecolor='none',
                        linewidth=2, alpha=0.8)

        ax.add_patch(circle)

        ax.text(centro_x - 0.1, # Posición x a la izquierda del punto (ajusta este valor)
                centro_y,
                str(i + 1),
                color=GRIS_5,
                fontsize=14,
                ha='right',
                va='center')

    # Configurar el título
    ax.set_title('Ventas totales mensuales en 2023', fontsize=18, color=GRIS_5)

    # Configurar el eje y
    def millions_formatter(x, pos):
        return f'MXN${x/1000000:.2f} M'
    ax.yaxis.set_major_formatter(FuncFormatter(millions_formatter))
    ax.tick_params(axis='y', labelsize=12, colors=AZUL_1)
    ax.set_ylabel('')

    # Configurar el eje x
    ax.set_xticks(range(len(ventas_mensuales_data.index)))
    ax.set_xticklabels(meses.values(), rotation=0, ha='center')
    ax.tick_params(axis='x', labelsize=12, colors=AZUL_1)
    ax.set_xlabel('')

    # Suprimir bordes del gráfico
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_color(GRIS_5)
    ax.spines['bottom'].set_color(GRIS_5)

    ax.grid(axis='y', linestyle='--', alpha=0.7, color=GRIS_3)

    plt.tight_layout()
    return fig # Devuelve el objeto figura


def facturacion_por_categoria():
    """Genera el gráfico de barras horizontales de facturación por categoría."""
    # Calcular la facturación para cada fila
    df['facturacion_producto'] = df['precio_unitario'] * df['cantidad']
    df['facturacion_total'] = df['facturacion_producto'] + df['envio']

    # Agrupar por categoría y sumar la facturación total
    facturacion_por_categoria_data = df.groupby('categoria')['facturacion_total'].sum().reset_index()

    # Ordenar por facturación descendente
    facturacion_por_categoria_data = facturacion_por_categoria_data.sort_values(by='facturacion_total', ascending=False)

    # Calcular el porcentaje de la primera barra
    if not facturacion_por_categoria_data.empty:
        valor_primera_barra = facturacion_por_categoria_data['facturacion_total'].iloc[0]
        suma_total_barras = facturacion_por_categoria_data['facturacion_total'].sum()
        porcentaje = (valor_primera_barra / suma_total_barras) * 100
    else:
        porcentaje = 0

    # Crear la visualización
    fig, ax = plt.subplots(figsize=(10, 6), facecolor=GRIS_1)
    ax.set_facecolor(GRIS_1)

    sns.barplot(x='facturacion_total', y='categoria', data=facturacion_por_categoria_data, color=AQUA_1, ax=ax)

    # Configurar el título
    ax.set_title('Facturación por categoría de producto en 2023', fontsize=18, color=GRIS_5)

    # Configurar el eje x
    ax.set_xticks([])
    ax.set_xlabel('')

    # Configurar el eje y
    ax.tick_params(axis='y', labelsize=12, colors=AZUL_1)
    ax.set_ylabel('')

    # Suprimir todos los bordes del gráfico
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)

    # Añadir el total de facturación al lado derecho de cada barra
    for p in ax.patches:
        width = p.get_width()
        plt.text(width + (ax.get_xlim()[1] * 0.01),
                 p.get_y() + p.get_height() / 2,
                 f'MXN${width:,.2f}',
                 ha='left',
                 va='center',
                 color=AZUL_1,
                 fontsize=12)

    # Agregar el texto dentro del área del gráfico en la posición especificada
    plt.text(0.4e7, 5,
             f"$\\bf{porcentaje:.2f}$% de las ventas corresponden únicamente a las ventas de Electrónica.\nEsto equivale a casi el doble en relación con todas las demás categorías\npresentes en el portafolio de Zoop.",
             fontsize=12, color=AZUL_3,
             ha='left', va='center', wrap=True)

    plt.tight_layout()
    return fig # Devuelve el objeto figura


def ventas_por_trimestre():
    """Genera el gráfico de barras apiladas de ventas por trimestre y método de pago."""
    # Asegurarse de que la columna 'fecha' esté en formato datetime
    df['fecha'] = pd.to_datetime(df['fecha'])

    # Calcular la facturación si aún no existe la columna 'facturacion'
    if 'facturacion' not in df.columns:
        df['facturacion'] = (df['precio_unitario'] * df['cantidad']) + df['envio']

    # Extraer el trimestre de la fecha
    df['trimestre'] = df['fecha'].dt.to_period('Q')

    # Agrupar por trimestre y método de pago, y sumar la facturación
    ventas_trimestre_metodo_data = df.groupby(['trimestre', 'metodo_pago'])['facturacion'].sum().reset_index()

    # Pivotear la tabla para que los métodos de pago sean columnas
    ventas_trimestre_metodo_pivot = ventas_trimestre_metodo_data.pivot(index='trimestre', columns='metodo_pago', values='facturacion').fillna(0)

    # Definir la paleta de colores para las barras apiladas
    palette_colores = [BLANCO, GRIS_3, GRIS_1, GRIS_3]
    # Nota: Es posible que necesites ajustar el orden de los colores

    # Crear el gráfico de barras apiladas
    fig, ax = plt.subplots(figsize=(10, 6), facecolor=GRIS_1)
    ax.set_facecolor(GRIS_1)

    ventas_trimestre_metodo_pivot.plot(kind='bar', stacked=True, ax=ax, color=palette_colores)

    # Configurar el título
    ax.set_title('Ventas por Trimestre y Método de Pago en 2023', fontsize=18, color=GRIS_5)

    # Configurar el eje x
    ax.tick_params(axis='x', rotation=0, labelsize=12, colors=AZUL_1)
    ax.set_xlabel('')

    # Configurar el eje y
    def millions_formatter_no_decimal(x, pos):
        return f'MXN${x/1000000:.0f} M'
    ax.yaxis.set_major_formatter(FuncFormatter(millions_formatter_no_decimal))
    ax.tick_params(axis='y', labelsize=12, colors=AZUL_1)
    ax.set_ylabel('')

    # Suprimir todos los bordes del gráfico
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)

    # Configurar la leyenda
    ax.legend(title='Método de Pago', loc='upper left', bbox_to_anchor=(1, 1), facecolor=GRIS_1, edgecolor=GRIS_5, title_fontsize='10', fontsize='8')

    # Agregar cuadrícula punteada en el eje Y
    ax.grid(axis='y', linestyle='dotted', color=GRIS_5, alpha=0.7)

    plt.tight_layout()
    return fig # Devuelve el objeto figura


def proporcion_cashback():
    """Genera el gráfico de dona de la proporción de clientes con cashback."""
    # Contar la frecuencia de cada valor en la columna 'cashback'
    conteo_cashback_data = df['cashback'].value_counts()

    # Definir los colores según los valores de 'cashback'
    colores_cashback = [GRIS_3, AZUL_1] # Asumiendo que conteo_cashback['No'] viene primero, luego conteo_cashback['Sí']

    # Crear la figura con el color de fondo especificado
    fig, ax = plt.subplots(figsize=(6, 6), facecolor=GRIS_1)

    # Configurar el fondo del gráfico (el área circular)
    ax.set_facecolor(GRIS_1)

    # Crear el gráfico de dona
    wedges, texts, autotexts = ax.pie(conteo_cashback_data,
                                       labels=conteo_cashback_data.index,
                                       autopct='%1.1f%%',
                                       startangle=180,
                                       colors=colores_cashback,
                                       wedgeprops=dict(width=0.4),
                                       pctdistance=0.8)

    # Configurar el color y tamaño de las etiquetas de texto (los labels 'Sí'/'No')
    for text in texts:
        text.set_color(text.get_color())
        text.set_fontsize(12)

    # Configurar el color y tamaño de los porcentajes
    for autotext in autotexts:
        autotext.set_color(BLANCO)
        autotext.set_fontsize(12)

    # Configurar el título de la figura
    fig.suptitle('¿Participan en el programa de cashback?',
                 fontsize=18,
                 color=GRIS_5)

    # Asegurar que el gráfico sea circular
    ax.axis('equal')

    plt.tight_layout()
    return fig # Devuelve el objeto figura


def distribucion_calificaciones():
    """Genera el histograma de la distribución de calificaciones."""
    # Crear la figura con el color de fondo especificado
    fig, ax = plt.subplots(figsize=(10, 6), facecolor=GRIS_1)
    ax.set_facecolor(GRIS_1)

    # Crear el histograma
    sns.histplot(data=df, x='nota', bins=11, kde=False, color=ROJO_1, ax=ax)

    # Configurar el título
    ax.set_title('Distribución de las evaluaciones de las compras en 2023', fontsize=18, color=GRIS_5)

    # Configurar el eje x
    ax.set_xticks(range(11))
    ax.tick_params(axis='x', labelsize=12, colors=AZUL_1)
    ax.set_xlabel('Calificación', fontsize=12, color=AZUL_1)

    # Configurar el eje y
    ax.tick_params(axis='y', labelsize=12, colors=AZUL_1)
    ax.set_ylabel('Frecuencia', fontsize=12, color=AZUL_1)

    # Suprimir todos los bordes del gráfico
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)

    # Agregar cuadrícula punteada en el eje Y detrás del gráfico
    ax.set_axisbelow(True)
    ax.grid(axis='y', linestyle='--', alpha=0.7, color=GRIS_3)

    # Calcular las estadísticas para la anotación
    promedio_calificaciones = df['nota'].mean().round(2)
    calificacion_mas_comun = df['nota'].value_counts().idxmax()
    frecuencia_mas_comun = df['nota'].value_counts().max()

    # Crear el texto de la anotación
    texto_anotacion = (
        f"Promedio de calificaciones = {promedio_calificaciones}\n"
        f"Calificación más común = {calificacion_mas_comun}\n"
        f"Frecuencia más común = {frecuencia_mas_comun}"
    )

    # Añadir la caja de anotación con texto y flecha
    ax.annotate(texto_anotacion,
                xy=(8.7, 2250),
                xytext=(2, 1800),
                fontsize=12,
                color=AZUL_1,
                bbox=dict(boxstyle="round,pad=0.5", fc=BLANCO, ec=GRIS_3, lw=1),
                arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=-0.2", color=AZUL_1, lw=1.5))

    plt.tight_layout()
    return fig # Devuelve el objeto figura


def box_plot_edad_sexo():
    """Genera el boxplot de la distribución de edad por sexo biológico."""
    # Crear la figura con el color de fondo especificado
    fig, ax = plt.subplots(figsize=(10, 6), facecolor=GRIS_1)
    ax.set_facecolor(GRIS_1)

    # Definir la paleta de colores para los boxplots
    palette_boxplot = [AZUL_1, AQUA_1]
    # Asegúrate de que el orden de los colores coincida con el orden de las categorías

    # Crear el boxplot
    sns.boxplot(x='sexo_biologico', y='edad', data=df, palette=palette_boxplot, ax=ax)

    # Configurar el título
    ax.set_title('Distribución de la Edad por Sexo Biológico de los clientes', fontsize=18, color=GRIS_5)

    # Configurar el eje x
    ax.set_xlabel('')
    ax.tick_params(axis='x', labelsize=12, colors=AZUL_1)

    # Configurar el eje y
    ax.tick_params(axis='y', labelsize=12, colors=AZUL_1)
    ax.set_ylabel('Edad', fontsize=12, color=AZUL_1)

    # Suprimir todos los bordes del gráfico
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)

    # Agregar cuadrícula punteada en el eje Y detrás del gráfico
    ax.set_axisbelow(True)
    ax.grid(axis='y', linestyle='--', alpha=0.7, color=GRIS_3)

    # Calcular las estadísticas para ambos grupos de sexo biológico
    edad_stats_por_sexo = df.groupby("sexo_biologico")["edad"]

    # Estadísticas para el primer grupo (índice [0])
    min_edad_grupo1 = edad_stats_por_sexo.min().iloc[0]
    q1_edad_grupo1 = edad_stats_por_sexo.quantile(.25).iloc[0]
    mediana_edad_grupo1 = edad_stats_por_sexo.median().iloc[0]
    q3_edad_grupo1 = edad_stats_por_sexo.quantile(.75).iloc[0]
    max_edad_grupo1 = edad_stats_por_sexo.max().iloc[0]

    # Estadísticas para el segundo grupo (índice [1])
    min_edad_grupo2 = edad_stats_por_sexo.min().iloc[1]
    q1_edad_grupo2 = edad_stats_por_sexo.quantile(.25).iloc[1]
    mediana_edad_grupo2 = edad_stats_por_sexo.median().iloc[1]
    q3_edad_grupo2 = edad_stats_por_sexo.quantile(.75).iloc[1]
    max_edad_grupo2 = edad_stats_por_sexo.max().iloc[1]

    # Crear el texto de la primera anotación (para el grupo con índice 1)
    texto_anotacion1 = (
        f"Mínimo = {min_edad_grupo2}\n"
        f"1º quartil = {q1_edad_grupo2}\n"
        f"2º quartil = {mediana_edad_grupo2}\n"
        f"3º quartil = {q3_edad_grupo2}\n"
        f"Máximo = {max_edad_grupo2}"
    )

    # Crear el texto de la segunda anotación (para el grupo con índice 0)
    texto_anotacion2 = (
        f"Mínimo = {min_edad_grupo1}\n"
        f"1º quartil = {q1_edad_grupo1}\n"
        f"2º quartil = {mediana_edad_grupo1}\n"
        f"3º quartil = {q3_edad_grupo1}\n"
        f"Máximo = {max_edad_grupo1}"
    )

    # Añadir la primera caja de anotación con texto y flecha
    ax.annotate(texto_anotacion1,
                xy=(-0.1, 45),
                xytext=(-0.4, 50),
                fontsize=12,
                color=AZUL_1,
                bbox=dict(boxstyle="round,pad=0.5", fc=BLANCO, ec=GRIS_3, lw=1),
                arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=-0.2", color=AZUL_1, lw=1.5))

    # Añadir la segunda caja de anotación con texto y flecha
    ax.annotate(texto_anotacion2,
                xy=(0.9, 45),
                xytext=(0.6, 50),
                fontsize=12,
                color=AZUL_1,
                bbox=dict(boxstyle="round,pad=0.5", fc=BLANCO, ec=GRIS_3, lw=1),
                arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=0.2", color=AZUL_1, lw=1.5))

    plt.tight_layout()
    return fig # Devuelve el objeto figura


# --- Exportar los gráficos ---

# Lista de funciones de gráficos
funciones_graficos = [
    metodo_pago,
    ventas_mensuales,
    facturacion_por_categoria,
    ventas_por_trimestre,
    proporcion_cashback,
    distribucion_calificaciones,
    box_plot_edad_sexo
]

# Exportar cada gráfico
for func in funciones_graficos:
    nombre_funcion = func.__name__
    nombre_imagen = f'grafico_{nombre_funcion}'

    # Generar el gráfico
    fig = func()

    # Guardar la figura en formato PNG
    fig.savefig(f'{nombre_imagen}.png', bbox_inches='tight')

    # Cerrar la figura para no mostrarla
    plt.close(fig)

print("Todos los gráficos han sido exportados como imágenes PNG.")
````
